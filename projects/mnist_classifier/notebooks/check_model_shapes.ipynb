## 2. Model shapes testing
We load the model, look at the data before running through the model and after. We throw an AssertionError if something doesn't add up.

```python
# Loading the model
model = MNISTClassifier()
```

```python
# Loading the loaders and getting the batch
train_loader, test_loader = get_mnist_loaders()
images, labels = next(iter(train_loader))

assert images.shape == (64, 1, 28, 28), f"Got shape: {images.shape}, expected: (64, 1, 28, 28)"
assert labels.shape == (64,), f"Got shape {labels.shape}, expected: (64,)"
print(f"Images shape: {images.shape}")
print(f"Labels shape: {labels.shape}")
```

```python
# Passing data through the model and getting results
output = model(images)

assert output.shape == (64, 10), f"Got shape {output.shape}, expected: (64, 10)"
print(f"Output shape: {output.shape}")
```
